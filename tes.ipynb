# Matplotlib & Pandas Basics

This notebook covers basics of **data reading using pandas** and **visualization using matplotlib**.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Show plots inside notebook
%matplotlib inline
```

## Reading Data from CSV

```python
df = pd.read_csv('sample_data.csv')
print(df)
```

## Exploring Data

```python
df.head()  # show first 5 rows

df.info()  # info about dataset

df.describe()  # summary statistics
```

## Line Plot

```python
# plt.plot(df['Name'], df['Score'], marker='o', color='blue')
plt.bar(df['Name'], df['Age'],color='green')




plt.title('Scores of Students')
plt.xlabel('Name')
plt.ylabel('Age')
# plt.grid(True)
plt.show()
```

## Bar Chart

```python
plt.bar(df['Name'], df['Age'], color='orange')
plt.title('Age of Students')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
```

## Scatter Plot

```python
plt.scatter(df['Age'], df['Score'], color='green')
plt.title('Score vs Age')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
```

## Histogram

```python
plt.hist(df['Score'], bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Scores')
plt.xlabel('Score Range')
plt.ylabel('Frequency')
plt.show()
```
