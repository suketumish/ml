# Getting Started With Pyplot

**Pyplot** is a Matplotlib module that provides simple functions for adding plot elements, such as lines, images, text, etc., to the axes in the current figure.  

Let's begin with a simple graph using fundamental Matplotlib code:

```python
fig, ax = plt.subplots()  # this creates a figure paired with axes


`````python
fig, ax = plt.subplots()
# x data
x = np.linspace(15, 10)
# y data
y = np.sin(x) + x
# creates the plot
ax.plot(x, y)                
# this will show your plot in case your environment is not interactive
fig.show()        
`````

`````python
color = 'r'
linestyle = 'dashed'
linewidth = 4
fig, ax = plt.subplots(figsize=(15,10))
fig.suptitle('Our First plot', fontsize=40)
ax.plot(x, y, c=color, linestyle=linestyle, linewidth=linewidth)
plt.show()
`````

`````python
ax.set_ylabel('function', fontsize = 30)
ax.set_xlabel('argument', fontsize = 30) 
`````

`````python
fig, ax = plt.subplots(figsize=(12,7))
fig.suptitle('Our First plot', fontsize=30)
ax.plot(x, y, c=color, linestyle=linestyle, linewidth=linewidth)
ax.set_ylabel('function', fontsize = 30)
ax.set_xlabel('argument', fontsize = 30)
for xtick in ax.xaxis.get_major_ticks():
    xtick.label1.set_fontsize(18)

for ytick in ax.yaxis.get_major_ticks():
    ytick.label1.set_fontsize(34)
plt.show()
`````

`````python
fig, axes = plt.subplots(1, 3, figsize=(16,8))
ax1, ax2, ax3 = axes
x = np.linspace(0,10,500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

ax1.set_ylabel('sin(x)', fontsize=30)
ax1.set_xlabel('x', fontsize=30)
ax2.set_ylabel('cos(x)', fontsize=30)
ax2.set_xlabel('x', fontsize=30)
ax3.set_ylabel('tan(x)', fontsize=30)
ax3.set_xlabel('x', fontsize=30)
fig.tight_layout(pad=2)

ax1.plot(x, y1, c='r', linewidth=4)
ax2.plot(x, y2, c='g', linewidth=4)
ax3.plot(x, y3, c='b', linewidth=4)
plt.show()
`````


## The Pie Chart

A **pie chart** is a circular plot that shows how data items relate to one another. The total area of the pie chart represents **100% of the data**, and the area of slices (also called **wedges**) represents the percentage of individual data parts. Pie charts are especially useful when we need to **compare parts of a whole**.

---

## Creating a Pie Chart

To create a pie chart using Matplotlib, you can use the `pie()` function:

```python
plt.pie(data)
````

Here, `data` is an array of values that you want to plot.

### Optional Parameters of `plt.pie()`

Some commonly used optional parameters include:

* `explode` â†’ allows separating slices of a pie chart
* `labels` â†’ a list of strings to label each corresponding slice
* `labeldistance` â†’ radial distance at which pie labels are drawn (default is 1.1)
* `colors` â†’ colors of the slices
* `shadow` â†’ adds a shadow effect to the pie chart
* `startangle` â†’ chooses the starting angle of the plot
* `wedgeprops` â†’ customize various slice parameters
* `radius` â†’ sets the radius of the circle (default is 1)

---



`````python
data=[33,52,12,17,42,48]
plt.pie(data)
plt.show()
`````

`````python
labels=['Python','C','C++','PHP','Java','Ruby']
plt.pie(data,labels=labels)
plt.show()
`````

`````python
plt.figure(figsize=(9, 7))
plt.pie(data)
plt.title('The results of the programming language survey', fontsize=14)
plt.legend(labels)
plt.show()
`````

`````python
explode = [0.0, 0.08, 0.0, 0.0,0.0,0.0]
plt.pie(data, explode=explode, labels=labels)
plt.show()
`````

# Colors and shadows
* To make our graph even more illustrative, we can pass a list of colors to the color parameter and add a shadow underneath by setting the shadow to True:

`````python
colors = ['saddlebrown', 'wheat', 'crimson', 'lightgrey','lightblue','darkblue']
plt.pie(data, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        shadow=True)
plt.show()
`````

## Display the value
* The autopct parameter displays the percent value on the slices. If autopct is %.2f, then for each pie slice, the format string is %.2f, where %is a special character that tells when to type the value, f sets the result to be a floating-point type, and the .2 sets a limit to only 2 digits after the point.



`````python
plt.pie(data, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        autopct='%.2f%%',    # here we also add the % sign 
        shadow=True)
plt.show()
`````

# Starting Angle in Pie Charts

By default, pie charts are plotted from the **x-axis**, and slices are drawn **counterclockwise**.  

The `startangle` parameter allows you to define the angle where you want the pie chart to start. The default angle is `0`.  

You can also control the drawing direction of the slices using the `counterclock` parameter, which is **True by default** (i.e., slices are drawn counterclockwise).  

---

## Example: Starting Slices at 90 Degrees

If we want the slices to start at **90 degrees**, we can set:

```python
plt.pie(data, startangle=90)


`````python
fig, axes = plt.subplots(1, 2, figsize=(10, 6))
ax1, ax2 = axes

ax1.pie(data, labels=labels, colors=colors, startangle=90)
ax2.pie(data, labels=labels, colors=colors, startangle=90, counterclock=False)
ax1.set_title('Starting the plot at 90Â°')
ax2.set_title('Plotting clockwise')
plt.show()
`````

# The Doughnut Chart

A **doughnut chart** is very similar to a pie chart. However, since a doughnut chart has an **opening at the center**, it makes the slices appear more like bars.  

Matplotlib does **not provide a dedicated method** for plotting doughnut charts.  
However, we can use the `wedgeprops` parameter in the `plt.pie()` function to define the **width of the wedges** and create a doughnut effect:

```python
plt.pie(data, wedgeprops={'width': 0.4})
